## Loading raw csv data

Data acquired from: https://www.cryptodatadownload.com/data/gemini/ on 28/09/2022

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('Gemini_BTCUSD_d.csv', skiprows=2, parse_dates=True, infer_datetime_format=True)
array = df.values # the array you are interested in

dates = np.flip(array[0::14,1])
values = np.flip(array[0::14,3])

# print(len(dates))
# print(dates[0::20])
xticks = dates[0::20]
print(type(dates[1]))

for i,value in enumerate(xticks):
    xticks = value.split()[0]
    # xticks = np.split(value)[0]

fig, axes = plt.subplots()

plt.rcParams["figure.figsize"] = (20,10)
plt.rcParams["figure.dpi"] = 100

axes.plot(dates,values)
plt.title('Value of Bitcoin in USD ($) since 2015')
plt.xlabel('Date')
plt.ylabel('Value($)')
plt.xticks(dates[0::20])
plt.show()
```
