```python
import numpy as np

class Particle:
    
    def __init__(self, x, y, vx, vy, radius, mass):

        # Defining attributes of the particle class
        self.r = np.array((x,y))
        self.v = np.array((vx,vy))
        self.radius = radius
        self.mass = mass
    
    # Define methods to call the various properties of a particle with decorators to allow easier calling
    
    # This is a getter method
    # can access x by particlename.x
    @property
    def x(self):
        return self.r[0]
    
    # This is a setter method
    # can change the value of x by particlename.x = ###
    @x.setter
    def x(self,value):
        self.r[0] = value
        
    @property
    def y(self):
        return self.r[1]
    
    @y.setter
    def y(self,value):
        self.r[1] = value
        
    @property
    def vx(self):
        return self.v[0]
    
    @vx.setter
    def vx(self,value):
        self.v[0] = value
        
    @property
    def vy(self):
        return self.v[1]
    
    @vy.setter
    def vy(self,value):
        self.v[1] = value
        
    
    # Now we define the methods which determine the functions of a particles
    # overlapping?
    
    def overlaps(self, other):
    """Does the circle of this Particle overlap that of other?"""

    return np.hypot(*(self.r - other.r)) < self.radius + other.radius
    
```
